10. __Add__ the additional column __"income_cat"__ with the following income categories:
- lowest 25% -> "Low"
- 25th to 50th percentile -> "Below_Average"
- 50th to 75th percentile -> "Above_Average"
- 75th to 95th percentile -> "High"
- Above 95th percentile -> "Very High"

```python
df["income_category"] = pd.qcut(df.median_income, q = [0, 0.25, 0.5, 0.75, 0.95, 1],
                                labels = ["Low", "Below_average", "Above_Average", "High", "Very High"])
```

### Feature Engineering

```python
label = df.median_house_value.copy()
label
```

```python
features = df.drop(columns=['median_house_value'])
features.info()
```

```python
# calculate the z-score of all of the columns which are of type float in order to acheive a standarized scale 
# for training the machine learning model
from scipy import stats
f1 = features.select_dtypes("float").apply(lambda x: stats.zscore(x))
f1
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
# Chekcing if features have been standarized

f1.agg(['mean','std'])
```

```python
# Get the dummy variables for ocean_proximity 

dummy = pd.get_dummies(features.ocean_proximity)
dummy
```

```python
features = pd.concat([f1, dummy, df.income_category], axis=1)
features
```

### Splitting data into training and test datasets

```python
from sklearn.model_selection import train_test_split

X_train, X_test = train_test_split(features, test_size = 0.2, stratify=features['income_category'])
```

```python
#Checking if training set and testing set have overalpping entries
def check(df1, df2):
    for i in df1.index:
        if i not in df2.index:
            continue
        else:
            return False
    return True
check(X_train, X_test)
```

```python
X_train.income_category.value_counts(normalize=True)
```

```python
X_test.income_category.value_counts(normalize=True)
```

```python
features.income_category.value_counts(normalize=True)
```

```python
# Dropping income category column from test and train data sets as it is not relavent for training the model
X_train.drop(columns=['income_category'], inplace=True)
X_test.drop(columns=['income_category'], inplace=True)
```

```python
# Getting the house prices for the house in the training and testing data sets
Y_train=label.loc[X_train.index]
Y_test=label.loc[X_test.index]
Y_train
```

### Training the ML model (Random Forest Regressor)

Using a random forest regressor model to capture the non-colinearity between location (longitude and latitude) and house prices

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest_reg = RandomForestRegressor(random_state = 42, n_estimators = 500,
                                   max_features= "sqrt", max_depth= 75, min_samples_split=2)
```

```python
forest_reg.fit(X_train, Y_train)
```

```python
forest_reg.score(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
predict = forest_reg.predict(X_train)
predict
```

```python
forest_mse = mean_squared_error(Y_train, predict)
forest_rmse = np.sqrt(forest_mse)
forest_rmse
```

### Evaluating the model on test dataset

```python
forest_reg.score(X_test, Y_test)
```

```python
pred = forest_reg.predict(X_test)
pred
```

```python
test_rmse = np.sqrt(mean_squared_error(pred,Y_test))
test_rmse
```

```python
compare = pd.DataFrame({'True_Value': Y_test, 'Prediction': pred})
compare
```

```python
mean_difference = compare.True_Value.sub(compare.Prediction).abs()
mean_difference.mean()
```

#### Feature Importance
Upon completing the model it is important to find out which features have the highest impact on house price

```python
feature_importance = pd.Series(forest_reg.feature_importances_, index=X_train.columns).sort_values(ascending=True)
feature_importance
```

```python
feature_importance.plot.barh(figsize=(12,8))
plt.show()
```
